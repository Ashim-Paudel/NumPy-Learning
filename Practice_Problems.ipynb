```python
def area_polygon(vertexes):
    ver_new = np.vstack((vertexes, vertexes[0])) #stack first row to last
    xy_0 = ver_new[:len(ver_new)-1, 0] #all x
    xy_1 = ver_new[1:, 1] #all y but y[0] at last
    yx_0 = ver_new[:len(ver_new)-1, 1] #all y
    yx_1 = ver_new[1:, 0] #all x but x[0] at last
    S_1 = xy_0.dot(xy_1) 
    S_2 = yx_0.dot(yx_1)
    area = abs(S_1 - S_2)/2
    return area
```

calculating area from this figure



```python
ver = np.array(((0,169.02), (30.55, 29.94), (226.02, 0), (139.37, 152.26), (159.97, 340.65)))
area_polygon(ver)
```
