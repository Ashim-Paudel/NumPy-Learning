## Reading and Writing an Array to a File
- two methods available
- `np.loadtxt` and `np.genfromtext` advanced format
- `np.save` and `np.load` for binary

```python
#np.save

dt = np.dtype({'names':['students', 'percentage'], 'formats':['S20', 'u1']})
data = [('ram', '90'), ('ashim', '92'), ('hari', '89')]

students_data = np.array(data, dtype=dt)
np.save('datas/students_data', students_data) #this np.save will save in binary with .npy extension
```

```python
np.save('datas/students_data.txt', students_data) #will not save in .txt format, rather will save with .npy extension
```

```python
#np.load

load_data = np.load('datas/students_data.npy')
load_data
```

#### Important! Since we have invalid data like - in weight and other columns better is to use `np.genfromtext` but also we can use converter argument to open such data file

`converters= {col_number: function }`

```python
# using parser to handel exception while converting into float:
def float_convert(s):
    try:
        return float(s)
    except ValueError:
        return -1
    else:
        pass
    
dt = np.dtype({'names':['ID','sex','DOB','height','weight','BP','VO2max'], 'formats':['U7', 'S1', 'S11', 'f4', 'f4','S6', 'f4']})
data = np.loadtxt('datas/students-population.txt',dtype=dt, comments='#', skiprows=9, converters={4: float_convert, 6:float_convert})
```

```python
#now to find average height of male and female:
weight_bool = data['weight']>0
```

```python
male_weight = data['weight'][weight_bool & sex_bool_array]
female_weight = data['weight'][weight_bool & ~sex_bool_array]
```

```python
#to plot male and female weight and height data:
plot_data = np.copy(data)
plot_data.sort(order='weight') #this will sort only new list in terms of weight
sex_bool = plot_data['sex'] == b'M'
valid_data = plot_data['weight']>0

m_w_plot = plot_data['weight'][valid_data & sex_bool_array]
m_h_plot = plot_data['height'][valid_data & sex_bool_array]

f_w_plot = plot_data['weight'][valid_data & ~sex_bool_array]
f_h_plot = plot_data['height'][valid_data & ~sex_bool_array]

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(m_w_plot, m_h_plot, '--',label='male', marker='$\u2642$', ms=12, lw=1)
ax.plot(f_w_plot, f_h_plot, ":", label='female', marker='$\u2640$', ms=12, lw=1)
ax.set_xlabel("weight (kg)")
ax.set_ylabel("height (m)")
ax.legend()
```

## np.genfromtxt
- `np.genfromtxt(fname, dtype, comments, skip_header, skip_footer, converters, usecols, unpack)` only `fname` is required argument

#### headers and footers
- two optional arguments insted of skiprows
- `skip_header` and `skip_footer` giving number of lines to skip


#### fixed width field:
- in `delimiters` value can be provided as a sequence of integers giving the widths of each field to be read in where the datacolumns do not have delimiters

#### missing values
- missing values are well handled by `np.genfromtxt`
- following data sets are returned in case of missing values:
| Data type | Default value |
| -- | -- |
| int | -1 |
| float | np.nan |
| bool | False |
| complex | np.nan + 0.j |

```python
#reading the file:
dt = np.dtype({'names':['sn', 'gender', 'time_black', 'time_color'], 'formats':['u1', 'S1', 'f4', 'f4']})
data = np.genfromtxt('datas/stroop.txt', dtype=dt, delimiter=',', skip_header=1)

n_words = 25 #25 words given to read
filter_data = data[np.isfinite(data['time_black']) & np.isfinite(data['time_color'])]
```

```python
data
```

```python
filter_data
```

```python
fb = filter_data['time_black'][filter_data['gender']==b'F']
fc = filter_data['time_color'][filter_data['gender']==b'F']
mb = filter_data['time_black'][filter_data['gender']==b'M']
mc = filter_data['time_color'][filter_data['gender']==b'M']
```

```python
fig,ax = plt.subplots(figsize=(6,5), edgecolor='w', facecolor='w')
ax.bar(["Male", "Female"], [mb.mean(), fb.mean()],color='.3', label="Black", width=-.15, align='edge')
ax.bar(["Male", "Female"], [mc.mean(), fc.mean()],color='c', label="Color", width=.15, align='edge')
ax.legend(loc='upper center')
ax.grid()
ax.set_xlabel("Gender")
ax.set_ylabel("Average time required (s)")
ax.set_title("Stroop effect in Male and Female")
plt.savefig('images/Stroop effect in Male and Female.png')
plt.show()
```

## np.savetxt
- The np.savetxt function saves a NumPy array as a text file. Its call signature is
- `np.savetxt(fname , X, fmt='%.18e', delimiter =' ',newline='\n', header='', footer='', comments ='# ')`
- `X` argument represents array to be stored
